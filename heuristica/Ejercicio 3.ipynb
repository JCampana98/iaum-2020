# Grilla de colores

Se tiene una grilla de cuadrados de 9×9, cada uno comienza pintado aleatoriamente de rojo o azul. 

<img src="grilla.jpg" width="200">
Se quiere pintar algunos cuadrados de manera que si se divide la grilla total en sub-grillas de 3x3 cada una de estas sub-grillas contenga cuadrados del mismo color pero el color de cada sub-grilla sea diferente que el de las subgrillas adyacentes.
<img src="grilla-solucion-1.jpg" width="200"><img src="grilla-solucion-2.jpg" width="200" >
Resolver el problema de manera que no sea necesario pintar un cuadrado más de una vez y se pinte la cantidad mínima de cuadrados




```python
import random
import numpy


grilla = [[random.choice(['rojo', 'azul']) for j in range(0,9)] for i in range(0,9)] 
grilla_bonita=numpy.array(grilla)

distribucion1 = [
                ['rojo','azul','rojo'],
                ['azul','rojo','azul'],
                ['rojo','azul','rojo']]
distribucion2 = [
                ['azul','rojo','azul'],
                ['rojo','azul','rojo'],
                ['azul','rojo','azul']]

parece1 = 0
parece2 = 0
total_pinceladas = 0

print(grilla_bonita)


def color_predominante(subset):
    cont_azul=0
    cont_rojo=0
    for fila in subset:
        for elemt in fila:
            if elemt == 'rojo':
                cont_rojo= cont_rojo+1
            if elemt == 'azul':
                cont_azul= cont_azul+1
    
    if cont_rojo > cont_azul:
        return 'rojo'
    else:
        return 'azul'
    
def pintar(subset,color,pinceladas):
    i = 0
    for fila in subset:
        j = 0
        for elemt in fila:  
            if elemt == 'rojo' and color == 'azul':
                subset[i, j] = 'azul'
                pinceladas = pinceladas +1
            if elemt == 'azul' and color == 'rojo':
                subset[i, j] = 'rojo'
                pinceladas = pinceladas +1
            j = j+1
        i = i+1
    return subset, pinceladas
        
subset1 = grilla_bonita[:3, :3]
print("Subset 1:")
print(subset1)
print(color_predominante(subset1))

subset2 = grilla_bonita[:3, 3:6]
print("Subset 2:")
print(subset2)
print(color_predominante(subset2))

subset3 = grilla_bonita[:3, 6:9]
print("Subset 3:")
print(subset3)
print(color_predominante(subset3))

subset4 = grilla_bonita[3:6, :3]
print("Subset 4:")
print(subset4)
print(color_predominante(subset4))

subset5 = grilla_bonita[3:6, 3:6]
print("Subset 5:")
print(subset5)
print(color_predominante(subset5))

subset6 = grilla_bonita[3:6, 6:9]
print("Subset 6:")
print(subset6)
print(color_predominante(subset6))

subset7 = grilla_bonita[6:9, :3]
print("Subset 7:")
print(subset7)
print(color_predominante(subset7))

subset8 = grilla_bonita[6:9, 3:6]
print("Subset 8:")
print(subset8)
print(color_predominante(subset8))

subset9 = grilla_bonita[6:9, 6:9]
print("Subset 9:")
print(subset9)
print(color_predominante(subset9))

"""
   Se crea una matriz con los colores predominantes de cada
   subset para su posterior uso.
"""
matriz_color = [
    [color_predominante(subset1),color_predominante(subset2),color_predominante(subset3)],
    [color_predominante(subset4),color_predominante(subset5),color_predominante(subset6)],
    [color_predominante(subset7),color_predominante(subset8),color_predominante(subset9)]]

print(matriz_color)

"""
   Chequea cada elemento de la matriz con cada elemento
   de las matrices de distribucion para ver a cual distribucion
   se parece mas la matriz color.
"""

i = 0
for fila in matriz_color:
        j = 0
        for elemt in fila:  
            if matriz_color[i][j] == distribucion1[i][j]:
                parece1 = parece1+1
            if matriz_color[i][j] == distribucion2[i][j]:
                parece2 = parece2+1
            j = j+1
        i = i+1    

"""
   Dependiendo de a que matriz se parece mas,
   es como se va a pintar la grilla.
   
"""

if parece1 > parece2:
    print("Se parece mas a la distribucion 1")
    subset5, total_pinceladas = pintar(subset5,'rojo', total_pinceladas)
    subset1, total_pinceladas = pintar(subset1,'rojo', total_pinceladas)
    subset3, total_pinceladas = pintar(subset3,'rojo', total_pinceladas)
    subset7, total_pinceladas = pintar(subset7,'rojo', total_pinceladas)
    subset9, total_pinceladas = pintar(subset9,'rojo', total_pinceladas)
    subset2, total_pinceladas = pintar(subset2,'azul', total_pinceladas)
    subset4, total_pinceladas = pintar(subset4,'azul', total_pinceladas)
    subset6, total_pinceladas = pintar(subset6,'azul', total_pinceladas)
    subset8, total_pinceladas = pintar(subset8,'azul', total_pinceladas)
if parece2 > parece1:
    print("Se parece mas a la distribucion 2")
    subset5, total_pinceladas = pintar(subset5,'azul', total_pinceladas) 
    subset1, total_pinceladas = pintar(subset1,'azul', total_pinceladas)
    subset3, total_pinceladas = pintar(subset3,'azul', total_pinceladas)
    subset7, total_pinceladas = pintar(subset7,'azul', total_pinceladas)
    subset9, total_pinceladas = pintar(subset9,'azul', total_pinceladas)
    subset2, total_pinceladas = pintar(subset2,'rojo', total_pinceladas)
    subset4, total_pinceladas = pintar(subset4,'rojo', total_pinceladas)
    subset6, total_pinceladas = pintar(subset6,'rojo', total_pinceladas)
    subset8, total_pinceladas = pintar(subset8,'rojo', total_pinceladas)

"""
   Falta volver a poner los subset como la grilla original.
"""

print("Se dieron un total de:", total_pinceladas, "pinceladas")



```
